```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_tabela = pd.read_csv('/content/Tabela1.csv', sep=";")
df_tabela
```

```python
(df_tabela.estado_civil=="solteiro").sum()

## Frequências absolutas em casados e solteiros
freq_absoltua_solteiro = np.sum(df_tabela.estado_civil=="solteiro")
freq_absoluta_casado = np.sum(df_tabela.estado_civil=="casado")
```

```python
## Frequência relativa de casados e solteiros
display(freq_absoltua_solteiro/(freq_absoltua_solteiro + freq_absoluta_casado))
display(freq_absoluta_casado/(freq_absoltua_solteiro + freq_absoluta_casado))
```

```python
df_tabela.grau_instrucao.unique()
```

```python
freq_absoluta_ef = np.sum(df_tabela.grau_instrucao=='ensino fundamental')
freq_absoluta_em = np.sum(df_tabela.grau_instrucao=='ensino médio')
freq_absoluta_es = np.sum(df_tabela.grau_instrucao=='ensino superior')

print(
    freq_absoluta_ef, "\n",
    freq_absoluta_em, "\n",
    freq_absoluta_es
)
```

```python
print("Frequência relativa ou contagem normalizada:\n", df_tabela.grau_instrucao.value_counts(normalize=True))
print("\n----------------\n")
print("Frequência absoluta ou contagem não normalizada:\n", df_tabela.grau_instrucao.value_counts())
```

```python
freq_absoluta_grau = df_tabela.grau_instrucao.value_counts()
```

```python
plt.  ylabel("Frequência absoluta Grau Ins.", size=15)
plt.xlabel("Grau de instrução", size=15)
plt.bar(freq_absoluta_grau.index, freq_absoluta_grau.values)
plt.show()
plt.savefig("Grau_instrucao.png")
#
```

```python
plt.pie(freq_absoluta_grau, labels=freq_absoluta_grau.index, autopct='%1.1f%%')
plt.show()
plt.savefig("Grau_instrucao_pie.png")
#
```
